# EDA and Preprocessing

This notebook continues from the data collection phase. The goal is to perfom some initial cleaning and exploratory data analysis to better understand the structure of each dataset before moving on to the main analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
sales = pd.read_csv("C:/Users/George/Desktop/data/Video Game Sales 2024.csv")
metacritic_games = pd.read_csv("C:/Users/George/Desktop/data/Metacritic's Best Games 2025.csv")
metacritic_reviews = pd.read_csv("C:/Users/George/Desktop/data/Metacritic's Best Reviews 2025.csv")
steam = pd.read_csv("C:/Users/George/Desktop/data/Steam Games Dataset March 2025(Cleaned).csv")
```

## 1. Sales Dataset - Cleaning and Summary Statistics

Upon reviewing the sales dataset, the 'img' column seems to only contain image file path strings without any analyticla values. It will be dropped before proceeding.

```python
sales = sales.drop(columns=['img'])
sales.head()
```

```python
sales.describe()
```

```python
missing_pct = (sales.isnull().sum() / len(sales) * 100).round(2)
missing_pct = missing_pct[missing_pct > 0].sort_values(ascending=False)
print(missing_pct)
```

### Observations

After transforming the missing values sum into percentages it is noticable that the missing values situation is a bit more severe than initially anticipated. 'critic_score' is absent for almost 90% of records, which is a significant limitation considering that it is one of the key variables in this project. Similarly, 'total_sales' is missing for about 70% of the dataset, meaning that sales-based analysis will be restricted to the approximately 18,900 games that have reported figures.

The 'last_update' column also carries a high missingness but it isn't really relevant to the analysis so it will probably be dropped in later preprocessing.

## 2. Merging the Metacritic Datasets

The two Metacritic datasets share a common 'id' column which can be used to combine them into a single dataset.

```python
metacritic = metacritic_games.merge(metacritic_reviews, on='id', how='left', suffixes=('_game', '_review'))
print(metacritic.shape)
metacritic.head()
```

### Observations

The merged dataset contains significantly more rows than 'metacritic_games' alone, as each game is repeated once per individual review entry. For the purpose of this project, the aggregated score column ('metascore', 'userscore') already present in 'metacritic_games' will be used directly for correlation analysis, making the full merged dataset more relevant for potential future work such as review text analysis.

## 3. Exploratory Data Analysis

The following visualizations are intended to provide an initial understanding of the data distributions and relationships relevant to the project hypothesis.

### 3.1 Top 10 Genres by Total Global Sales

```python
genre_sales = sales.dropna(subset=['total_sales']).groupby('genre')['total_sales'].sum().sort_values(ascending=False).head(10)

genre_sales.plot(kind='bar', color='steelblue')
plt.title('Top 10 Genres by Total Global Sales')
plt.xlabel('Genre')
plt.ylabel('Total Sales (millions)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

print(genre_sales)
```

### Observations

Action and Shooter genres lead significantly in total global sales, which aligns with the commercial dominance of major franchises such as Grand Theft Auto and Call of Duty. This provides early support for the hypothesis that genre is a meaningful predictor of commecial success.
